```typescript
import { delay } from "https://deno.land/std/async/delay.ts";
```

```typescript
let pool = [];
for (let index = 0; index < 10; index++) {
    const response = await fetch(`https://api.geckoterminal.com/api/v2/networks/cro/pools?page=${index+1}`);
    const result = await response.json();
    pool = [...pool,...result.data];
    await delay(1000);
}
```

```typescript
pool[3]
```

```typescript
const select = pool.map(({attributes}) => {return({name:attributes.name,address:attributes.address})});
```

```typescript
await Deno.writeTextFile("src/pools.ts", `export const pools = ${JSON.stringify(select, null, 2).replaceAll('"name"','name').replaceAll('"address"','address')}`);
```
